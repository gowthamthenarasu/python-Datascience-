# ploting bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes =["python","R","AI","ML","DS"]
classes1_Student=[30,10,20,25,10]
classes2_Student=[40,5,20,25,10]
classes3_Student=[35,5,30,15,15]
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(classes, classes1_Student,width=0.4,align="edge",color="y",edgecolor="m",linewidth=5,alpha=0.9
       ,linestyle="--",label="classes1_Student",visible=True,)
```

```python
plt.figure(figsize=(16,9))


plt.bar(classes, classes1_Student,width=0.4,align="edge",color="y",edgecolor="m",linewidth=5,alpha=0.9
       ,linestyle="--",label="classes1_Student",visible=True,)
plt.title("Bar chart of IAIP class")
plt.xlabel("classes")
plt.ylabel("No of students")
plt.show()
```

```python
plt.figure(figsize=(5,9))
classes_index = np.arange(len(classes))

width=0.2

plt.bar(classes_index, classes1_Student,width,align="edge",color="y",edgecolor="m",linewidth=2,alpha=0.9
       ,linestyle="--",label="classes1_Student")#visible=True

plt.bar(classes_index + width, classes2_Student,align="edge",color="g",edgecolor="m",linewidth=2,alpha=0.4
       ,linestyle="--",label="classes2_Student")

plt.xticks(classes_index,classes)
plt.title("Bar chart of IAIP class",fontsize=10)
plt.xlabel("classes",fontsize=15)
plt.ylabel("No of students",fontsize=15)
plt.show()
```
